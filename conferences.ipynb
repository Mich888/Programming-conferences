```python
import csv

with open('konferencje.csv') as csvfile:
    konf = list(csv.DictReader(csvfile, delimiter = ';')) # wczytujemy plik jako listę słowników
```

```python
len(konf) # liczba wierszy w pliku
```

```python
ileRazy = {'Ruby': 0, 'Java': 0, 'C': 0, 'Python': 0, 'JavaScript': 0, 'Go': 0, 'C++': 0} # mapa zliczająca wartości
# dla danego języka
```

```python
for value in konf[0].keys(): # przechodzimy po nazwach konferencji
    if 'Ruby' in value:
        ileRazy['Ruby'] += 1
    if 'Java' in value and 'JavaScript' not in value:
        ileRazy['Java'] += 1
    if 'C' in value and 'C++' not in value:
        ileRazy['C'] += 1
    if 'Python' in value:
        ileRazy['Python'] += 1
    if 'JavaScript' in value:
        ileRazy['JavaScript'] += 1
    if 'Go' in value:
        ileRazy['Go'] += 1
    if 'C++' in value:
        ileRazy['C++'] += 1 
```

```python
for value in ileRazy:
    print(value + ": " + str(ileRazy[value]))
```

```python
for jezyk, liczba in ileRazy.items(): 
    print(jezyk + ": " + str(liczba))
```

```python
iluRóżnychUczestników = set()
for row in konf:
    for uczestnik in row.keys():
        if 'Python' in uczestnik:
            if row[uczestnik]:
                iluRóżnychUczestników.add(row[uczestnik].split('/')[0])
print(len(iluRóżnychUczestników))
iluRóżnychUczestników
```

```python
iluRóżnychUczestników2 = set(value for row in konf for value in row.values())
print(len(iluRóżnychUczestników2))
```

```python
for jezyk in ileRazy:
    ileRazy[jezyk] = 0
```

```python
for wiersz in konf:
    for klucz in wiersz.keys():
        start = '('
        end = ')'
        s = klucz
        jezyk = s[s.find(start)+len(start):s.rfind(end)]
        if wiersz[klucz]: # sprawdzenie czy string nie jest pusty
            ileRazy[jezyk] += int(wiersz[klucz].split('/')[1])
```

```python
max(ileRazy, key = ileRazy.get) # klucz z największą wartością w mapie
```

```python
# dodanie krajów do mapy
start = '.'
end = '/'
kraje = {}
for wiersz in konf:
    for wartość in wiersz.values():
        if wartość:
            s = wartość
            kraj = s[s.find(start)+len(start):s.rfind(end)]
            if kraj in kraje.keys():
                kraje[kraj] += int(wartość.split('/')[1])
            else:
                kraje[kraj] = 0;
```

```python
for kraj, liczba in kraje.items():
    print(kraj + ": " + str(liczba))
```

```python
# wpisanie wartości do mapy

start = '.'
end = '/'
for wiersz in konf:
    for wartość in wiersz.values():
        s = wartość
        kraj = s[s.find(start)+len(start):s.rfind(end)]
        if kraj: # sprawdzenie czy string nie jest pusty
            kraje[kraj] += int(wartość.split('/')[1])
```

```python
for kraj, liczba in kraje.items():
    print(kraj + ": " + str(liczba))
```

```python
max(kraje, key = kraje.get) # klucz z największą wartością w mapie
```
